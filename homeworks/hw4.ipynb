### Question 2

```python
# Probability error
d = 0.05
e = 0.05

# VC dimension
v = 50

# 100 linearly spaced numbers
x = np.linspace(0,10000,100)

# Original VC bound
def ya(x):
    y = np.sqrt((8/x)*np.log((4*(2*x)**v)/d))
    return y

# Rademacher Penalty Bound:
def yb(x):
    y = np.sqrt((2*np.log(2*x*(x**v)))/x) + np.sqrt((2/x)*np.log(1/d)) + 1/x
    return y

# Parrondo and Van den Broek
def yc(x):
    y = np.sqrt((1/x)*(2*e + np.log((6*(2*x)**v)/d)))
    return y

# Devroye
def yd(x):
    y = np.sqrt( (1/(2*x)) * ( 4*e*(1+e) + 2 * np.log(2*(x**v) / np.sqrt(d)) ) )
    return y

z = 0.05 + 0*x


# plot the function

plt.plot(x,ya(x), 'r')
plt.plot(x,yb(x), 'b')
plt.plot(x,yc(x), 'm')
plt.plot(x,yd(x), 'y')
plt.plot(x,z, 'g')

plt.show()

N = 10000.0
[ya(N), yb(N), yc(N), yd(N)]
```

✅ closest is answer [D]
