# Problem 1 - Generalization bound

### Question 1

functions of the form f(x)=klogx have attractive _fixed points_

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import math
```

```python
def fixed_point(f, x = 1, error = 0.001):
    while abs(x - f(x)) > error:
        x = f(x)
    return x

def f(x, epsilon = 0.05, delta = 0.05, d_vc = 10):
    return (8/epsilon**2)*math.log((4*(2*x)**d_vc)/delta)
```

```python
fixed_point(f)
```

✅ closest answer is 460 000, answer [D]

### Question 2

```python
# Probability error
delta = 0.05
epsilon = 0.05

# VC dimension
d_vc = 50

# 100 linearly spaced numbers
x = np.linspace(0,10000,100)

# Original VC bound
def ya(x):
    y = np.sqrt((8/x)*np.log((4*(2*x)**d_vc)/delta))
    return y

# Rademacher Penalty Bound:
def yb(x):
    y = np.sqrt((2*np.log(2*x*(x**d_vc)))/x) + np.sqrt((2/x)*np.log(1/delta)) + 1/x
    return y

# Parrondo and Van den Broek
def yc(x):
    y = np.sqrt((1/x)*(2*epsilon + np.log((6*(2*x)**d_vc)/delta)))
    return y

# Devroye
def yd(x):
    y = np.sqrt( (1/(2*x)) * ( 4*epsilon*(1+epsilon) + 2 * np.log(2*(x**d_vc) / np.sqrt(delta)) ) )
    return y

z = 0.05 + 0*x


# plot the function

plt.plot(x,ya(x), 'r')
plt.plot(x,yb(x), 'b')
plt.plot(x,yc(x), 'm')
plt.plot(x,yd(x), 'y')
plt.plot(x,z, 'g')

plt.show()

N = 10000.0
[ya(N), yb(N), yc(N), yd(N)]
```

✅ the smallest bound is Devroye, answer [D]

```python
def y(x):
    y = np.sin(np.pi * x)
    return y

def create_dataset():
    x1, x2 = np.random.uniform(-1,1,2)
    return [(x1, y(x1)),(x2, y(x2))]

a,b = create_dataset()
```
